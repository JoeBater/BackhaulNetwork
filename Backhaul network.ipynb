```python
k = 5

# move 
def move(nodes, n1, n2):
    
    # don't allow move to self
    if n1 == n2:
        print("move rejected - can't move to self")
        pass
    
    # ensure root node can't move
    elif n1 == 0:
        print("move rejected - root can't move")
        pass
    
    # otherwise
    else:
    
        # prevent cycling
        if n1 in nodes.loc[n2, 'path']:
            print('move rejected - introduces a cycle')
            
        else:
                
            # make a copy of the current nodes
            newnodes = nodes.copy()

            # set the connects value
            newnodes.loc[n1, 'connects'] = n2


            # set the path
            new_p = [n2] + newnodes.loc[n2, 'path'].copy()

            print(newnodes.loc[n1, 'path'], new_p)

            newnodes.iat[n1, 3] = new_p


            reject = 0

            # look through all paths that might contain n2
            for i in range(len(newnodes)):
                if n1 in newnodes.loc[i, 'path']:
                    p = newnodes.loc[i, 'path']
                    x = p.index(n1)+1
                    newnodes.iat[i, 3] = p[:x] + new_p

                if len(newnodes.loc[i, 'path']) > k:
                    reject = 1

            if reject == 1:
                print('move rejected - branch becomes too long')
            else:
                nodes = newnodes
                
    return nodes
```
